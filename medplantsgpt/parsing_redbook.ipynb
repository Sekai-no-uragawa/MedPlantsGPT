```python
for name, url, _ in list_of_links:
    if url:
        page = requests.get(url)
        if page.status_code == 200:
            soup = BeautifulSoup(page.text, "html.parser")
            allTags = soup.findAll('p')
            plants_redbooks[name] = []
            for tag in allTags:
                try:
                    if tag.find('a').get('title') == 'Смотреть список видов этой Красной книги':
                        link = tag.find('a').get('href')
                        region = tag.text
                        plants_redbooks[name].append((region, link))
                except:
                    pass
        else:
            
            print(f'Ошибка соединения с {url, name}')
```

```python
plants_redbooks
```

```python
clear_plants_redbooks = {}
for key, value in plants_redbooks.items():
    new_val = []
    for val in value:
        new_val.append((val[0].split('(')[0].strip(), val[1]))
    clear_plants_redbooks[key] = new_val
```

```python
clear_plants_redbooks
```

```python
results = {}
for key, value in clear_plants_redbooks.items():
    filtered_data = {}
    for item in value:
        region, link = item
        filtered_data[region] = link

    results[key] = list(filtered_data.items())
```
